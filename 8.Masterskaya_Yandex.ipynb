#### `gender`

Заметим, что в столбце `gender` только два значения (1 и 2), которые соответствуют полу. Приведем их к стандартным обозначениям (0 и 1), чтобы алгоритм не считал один признак значимее другого.

В столбце `age` возраст, предположительно, указан в днях. Переведем его в года.

```python
def height_treatment(df):
    for i in range(len(df)):
        if df['height'].iloc[i] < 100:
            df['height'].iloc[i] = int(str(1)+str(df['height'].iloc[i]))
    return df
```

#### `ap_hi` и `ap_lo`

Там, где ap_lo больше, чем ap_hi, необходимо поменять их местами.

```python
# index6 = test.loc[test['ap_lo'] > test['ap_hi']].index
# test.loc[test['ap_lo'] > test['ap_hi']]
```

```python
def ap_hi_lo_swap(df):
    for i in range(len(df)):
        if df['ap_lo'].iloc[i] > df['ap_hi'].iloc[i]:
            hi = df['ap_hi'].iloc[i].copy()
            lo = df['ap_lo'].iloc[i].copy()
            df['ap_lo'].iloc[i] = hi
            df['ap_hi'].iloc[i] = lo
    return df
```

Данные для таблицы взяты из **[данной статьи](https://ria.ru/20220514/ves-1788536975.html?ysclid=leh4vfqvz8710306160)**.

```python
data_bmi = {'ИМТ': ['< 16', '16—18,5', '18,5—24,9', '25—29,9', '30—34,9', 
             '35—39,9', '> 40'], 
     'Результат': ['Дефицит массы тела', 'Недостаточная масса тела', 'Норма',
                   'Лишний вес', 'Ожирение 1 степени', 'Ожирение 2 степени', 
                   'Ожирение 3 степени']}
bmi_table = pd.DataFrame(data=data_bmi)
display(bmi_table)
```

```python
# Категоризация ИМТ (BMI)
def categorize_bmi(bmi):
    if bmi < 16:
        return 1
    if 16 <= bmi < 18.5:
        return 2
    if 18.5 <= bmi <= 24.9:
        return 3
    if 25 <= bmi <= 29.9:
        return 4
    if 30 <= bmi <= 34.9:
        return 5
    if 35 <= bmi <= 39.9:
        return 6
    if bmi >= 40:
        return 7
```

```python
train['bmi_category'] = train['bmi'].apply(categorize_bmi)
test['bmi_category'] = test['bmi'].apply(categorize_bmi)
```

Удалим столбец `bmi`, чтобы не было мультиколлинеарности.

```python
train = train.drop(['bmi'], axis=1)
test = test.drop(['bmi'], axis=1)
```

#### Категоризация артериальной гипертензии (АГ)

Данные для таблицы взяты из **[Википедии](https://ru.wikipedia.org/wiki/Артериальная_гипертензия)**.

```python
data_hypertension = {'Категории АД': ['Оптимальное', 'Нормальное', 
                                'Высокое нормальное', 'АГ 1-й степени', 
                                'АГ 2-й степени', 'Изолированная систолическая АГ'], 
                     'САД': ['< 120', '< 130', '130—139', '140—159', '≥ 160', '≥ 140'], 
                     'ДАД': ['< 80', '< 85', '85—89', '90—99', '≥ 100', '< 90']}
hypertension_table = pd.DataFrame(data=data_hypertension)
display(hypertension_table)
```

Для тех случаев, когда показания САД и ДАД не подходят ни под одну из категорий, введем еще одну категорию (7) для всех таких показаний.

```python
def hypertension(ap_hi, ap_lo):
    if ap_hi < 120 and ap_lo < 80:
        return 1
    if 120 <= ap_hi < 130 and 80 <= ap_lo < 85:
        return 2
    if 130 <= ap_hi < 140 and 85 <= ap_lo < 90:
        return 3
    if 140 <= ap_hi < 160 and 90 <= ap_lo < 100:
        return 4
    if ap_hi >= 160 and ap_lo >= 100:
        return 5
    if ap_hi >= 140 and ap_lo < 90:
        return 6
    return 7
```

```python
train['hypertension'] = train.apply(lambda x: hypertension(x.ap_hi, x.ap_lo), axis=1)
test['hypertension'] = test.apply(lambda x: hypertension(x.ap_hi, x.ap_lo), axis=1)
```

#### Модель решающего дерева

- `max_depth = range(1, 11)`
- `max_features = range(1, len(features.columns)+1)`
